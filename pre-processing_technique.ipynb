## Now lets see the stop word removal techniques

```python
var="Natural Language Processing (NLP) enables machines to understand. human language. It powers chatbots, translation, and sentiment analysis. By preprocessing text—like. tokenization, lowercasing, and removing. stopwords—we make. data ready for AI models to .analyze and generate. meaningful insights. efficiently"
```

```python
var
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
stop=stopwords.words("english")
stop
```

```python
stopword_list=list(punctuation)+stop
```

```python
stopword_list
```

```python
for i in word:
    if i not in stopword_list:
        print(i)
```
