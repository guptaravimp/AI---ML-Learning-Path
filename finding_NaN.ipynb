## Now lets see How to fill missing content 

```python
dataset = pd.read_csv("dummy_loan_data_with_nan.csv")
print(dataset)
```

```python
dataset.head(4)
```

### to fill the data first we have to Know about the datatype of that column 

```python
dataset.isnull().sum()
```

```python
dataset.fillna(10) ## that fill all nan value with 10 but it fill nam emeail all the thing with 10 that is rong ways 
```

```python
# dataset.fillna(method="ffill",axis=1)  ## if axis=1 means column wise data filling and if axis=0 means row wise data filling 
dataset.bfill(axis=1)  # Backward fill column-wise
```

```python
dataset.info()
```

```python
dataset.ffill(axis=1)
```

```python
# mod filling is that we fill the null data with nmaximum repeating val;ue in the file ?
# that is in specific column 
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
#now filling all the Nan with mod value 
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

## 3. Now let see handling missisng value using Scikit-Learn

```python
dataset = pd.read_csv("dummy_loan_data_with_nan.csv")
print(dataset)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="mean")
ar=si.fit_transform(dataset[['Dependents', 'Credit_History']])
```

```python
new_dataset=pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```

```python
print(new_dataset)
```

```python
new_dataset.isnull().sum()
```
